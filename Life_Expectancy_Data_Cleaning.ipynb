```python
def headers_to_lowercase (df):

    df.columns = df.columns.str.lower()
    
    return df
```

```python
def text_to_lowercase (df):
    
    df = df.applymap(lambda s: s.lower() if type(s) == str else s)
    
    return df   
```

```python
life_df = headers_to_lowercase(life_df)
life_df = text_to_lowercase(life_df)
```

```python
def drop_columns (df, list_of_columns_to_drop):
    df.drop(columns = list_of_column_to_drop, inplace=True)
    return df

list_of_columns_to_drop = ['country_code', 'hospitals','gni_capita','une_hiv', 'une_poverty', 'une_edu_spend', 'une_literacy',
       'une_school', 'une_infant', 'une_life']    
```

```python
## creates a list of countries which will be used in the drop function
country_drop = ['montenegro','south sudan',"democratic people’s republic of korea", 'albania', 'bosnia and herzegovina','syrian arab republic','cuba','djibouti']
## creates a function that drops based on unwanted records/countries

def row_drop (df, column_title, country_drop):
    df = df[~df[column_title].isin(country_drop)]
    return df

life_df = row_drop(life_df, 'country', country_drop)
```

```python
singapore_pop_dict = {2000:4028.871 , 2001:4077.145 , 2002:4104.419 , 2003:4129.225 , 2004: 4177.213 , 2005: 4265.687 , 2006:4402.331 , 2007:4578.630 , 2008:4775.804 , 2009:4966.614 , 2010:5131.172 , 2011:5263.636, 2012:5369.476, 2013:5453.737, 2014:5525.626, 2015:5592.152, 2016:5653.634}
ukraine_pop_dics = {2014:45111.532, 2015:44921.639, 2016: 44713.702}
brazil_pop_dict = {2000:174790.340, 2006: 188167.356, 2010:195713.635}
```

```python
def fill_na_in_population_with_values_from_dict (df):
    
    singapore_pop_dict = {2000:4028.871 , 2001:4077.145 , 2002:4104.419 , 2003:4129.225 , 2004: 4177.213 , 2005: 4265.687 , 2006:4402.331 , 2007:4578.630 , 2008:4775.804 , 2009:4966.614 , 2010:5131.172 , 2011:5263.636, 2012:5369.476, 2013:5453.737, 2014:5525.626, 2015:5592.152, 2016:5653.634}
    ukraine_pop_dict = {2014:45111.532, 2015:44921.639, 2016: 44713.702}
    brazil_pop_dict = {2000:174790.340, 2006: 188167.356, 2010:195713.635}
    
    life_df['une_pop'][life_df['country']=="singapore"] = life_df['une_pop'][life_df['country']=="singapore"].fillna(life_df['year'].map(singapore_pop_dict))
    
    life_df['une_pop'][life_df['country']=="brazil"] = life_df['une_pop'][life_df['country']=="brazil"].fillna(life_df['year'].map(brazil_pop_dict))
    
    life_df['une_pop'][life_df['country']=="ukraine"] = life_df['une_pop'][life_df['country']=="ukraine"].fillna(life_df['year'].map(ukraine_pop_dict))
    
    return df
```

```python
def change_col_names (df):
    
    column_name_dict = {'life_expect':'life_expectancy','life_exp60':'life_expectancy_at_60','infant_mort':'infant_mortality','age1-4mort':'age_1-4_mortality', 'age5-19thinness':'age_5-19_thinness', 'age5-19obesity':'age_5-19_obesity', 'hepatitis':'hepatitis_vacc', 'measles':'measles_vacc','polio':'polio_vacc', 'diphtheria':'diphteria_vacc', 'basic_water':'basic_water_access', 'doctors':'doctors_per_10000', 'une_pop':'population', 'une_gni':'gni'}
    df.rename(columns=column_name_dict, inplace= True)
    return df
```

```python
change_col_names(life_df)
```

```python
def replace_category_mean (df):

    df['alcohol'] = df['alcohol'].fillna(df.groupby('country')['alcohol'].transform('mean'))
    
    df['hepatitis_vacc'] = df['hepatitis_vacc'].fillna(df.groupby('country')['hepatitis_vacc'].transform('mean'))
    
    df['doctors_per_10000'] = df['doctors_per_10000'].fillna(df.groupby('country')['doctors_per_10000'].transform('mean'))
    
    df['gni'] = df['gni'].fillna(df.groupby('country')['gni'].transform('mean'))
    
    return df
```

```python
replace_category_mean(life_df)
```
